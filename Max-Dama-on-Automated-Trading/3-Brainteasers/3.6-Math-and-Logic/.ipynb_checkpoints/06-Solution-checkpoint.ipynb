# A line segment is broken into 3 pieces. What is the probability that they form a triangle?

___
# Solution

- First, we have to think about the conditions for a line segment to form a triangle

![](https://upload.wikimedia.org/wikipedia/commons/3/32/Trigo-unitcircle-animation.gif)

- The gif shows that **the length of the longest face of the triangle must always be less than the sum of the length of the other two**
    - WHY IS THIS TRUE?
        - Imagine the length of one side was longer than the sum of the other two
            - Then, even if you layed all pieces flat, you wouldn't be able to reach from one end of the longest piece to the other to close the triangle

- So how can we formalize these conditions
    - We break the line segment into three pieces - let's call them $a, b, \text{and } c$. Then:
    
# $ a < b + c$
# $b < a + c$
# $ c < a + b$

- But since we're breaking the line segment into three pieces, if we know $a$ and $b$ then we know the length of $c$ is simply what's left of the line segment. This means that we can express $c$ as:

# $c = L - (a + b)$

- Where $L$ is the total length of the line segment being broken up

- Now, we can sub this expression for $c$ into the conditions:

# $ a < b + L - (a + b) \implies a < L - a \implies a < L/2$
# $b < a + L - (a + b) \implies b < L - b \implies b < L/2$
# $ L - (a + b) < a + b \implies L < 2(a + b) \implies a + b > L/2 $

- So, expressing these conditions in plain language, we know:
    - (1) The length of parts $a$ and $b$ must make up more than half of the total length
    - (2) Neither $a$ nor $b$ **alone** can have length longer than half the total length

- We can start to think about these conditions with a plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5

x = range(11)
y = [10-i for i in x]

plt.plot(x,y)
plt.plot([float(i)/2 for i in x],[float(i)/2 for i in y])

plt.axvline(5, ls = 'dotted', color = 'r')
plt.axhline(5, ls = 'dotted', color = 'r')

my_ticks = ['0','','','','','L/2','','','','','L']
plt.xticks(x, my_ticks)
plt.yticks(x, my_ticks)
plt.xlim([0,10])
plt.ylim([0,10])
plt.xlabel('Length of a')
plt.ylabel('Length of b')

plt.show()
```

- So what is this plot showing us?
    - It represents different combinations of lengths $a$ and $b$
    - First, notice that **the x axis represents the length of piece $a$, while the y axis represents the length of piece $b$**
    
- Now, what can we say about the functions?


1. The blue line represents the function $x+y=L$ (or alternatively $y = L-x$)
    - So first, we realize that anything above (i.e. greater) than this blue function cannot exist under our conditions
        - **WHY?**: because it would mean that the value of $a + b$ is greater than $L$
            - *How can you break a line segment of length $L$ into pieces whose sum is greater than $L$?*
            
2. The blue line itself represents combinations that cannot exist under our conditions
    - **WHY?**: because the line represents combinations where the sum of $a$ and $b$ are equal to $L$
        - What about piece $c$? This creates only two pieces, instead of three
            - Therefore, we can conclude all possible combinations must be below the blue function
            
3. The yellow line represents the function $x+y=L/2$ (or alternatively $y = L/2-x$)
    - We know from our conditions above that $a + b$ must be greater than $L/2$
        - Therefore, no possible combinations can exist below the yellow line
        
- Let's take a break and summarize the regions that are off limits based on our conclusions

```python
x = range(11)
y = [10-i for i in x]

plt.plot(x,y, 'k')
plt.plot([float(i)/2 for i in x],[float(i)/2 for i in y], 'k', alpha = 0.5)

plt.axvline(5, ls = 'dotted', color = 'r')
plt.axhline(5, ls = 'dotted', color = 'r')

plt.fill_between([float(i)/2 for i in x], 0, [float(i)/2 for i in y], color = 'k', alpha = 0.5)
plt.fill_between(x, y, 10, color = 'k')

my_ticks = ['0','','','','','L/2','','','','','L']
plt.xticks(x, my_ticks)
plt.yticks(x, my_ticks)
plt.xlim([0,10])
plt.ylim([0,10])
plt.xlabel('Length of a')
plt.ylabel('Length of b')

plt.show()
```

- Continuing on, the horizontal dotted line represents the threshold between lengths of $b$ that are greater than/less than $L/2$
    - **One of our conditions is that $b$ must be less than $L/2$**
        - Therefore, any area above the dotted horizontal line violates the conditions
        
- Similarly, anything to the right of the vertical line violates the conditions

Therefore, our plot becomes:

```python
x = range(11)
y = [10-i for i in x]

plt.plot(x,y, 'k')
plt.plot([float(i)/2 for i in x],[float(i)/2 for i in y], 'k', alpha = 0.5)

plt.axvline(5, ls = 'dotted', color = 'r')
plt.axhline(5, ls = 'dotted', color = 'r')

plt.fill_between([float(i)/2 for i in x], 0, [float(i)/2 for i in y], color = 'k', alpha = 0.5)
plt.fill_between(x, y, 10, color = 'k')

plt.fill_between([float(i)/2 for i in x], 5, 10, color = 'k', alpha = 0.5)
plt.fill_between([5+float(i)/2 for i in x], 0, 5, color = 'k', alpha = 0.5)

my_ticks = ['0','','','','','L/2','','','','','L']
plt.xticks(x, my_ticks)
plt.yticks(x, my_ticks)
plt.xlim([0,10])
plt.ylim([0,10])
plt.xlabel('Length of a')
plt.ylabel('Length of b')

plt.show()
```

- In summary, the non-black region of the chart represents all possible combinations of breaking the line segment
    - The white region, however, represents the **combinations that satisfy our conditions**
    
- This white region makes up 1/4 of the possible combinations

# Therefore, P(The three pieces forming a triangle) = 1/4

# Simulation solution

- What if we randomly chopped up the line segment many times, then tested to see if it could for a triangle?

```python
import random

def simulation():
    #Assume the length of the line segment is 1
    #Generate two random points between 0 and 1 to simulate the break points for the line segment
    break_1 = random.random()
    break_2 = random.random()
    
    #using these break points, we can define lengths a, b, and c
    min_break = min(break_1, break_2)
    max_break = max(break_1, break_2)
    
    a = min_break
    b = max_break - a
    c = 1 - (a+b)
    
    #now, we can test our conditions:
    condition_1 = a < 0.5
    condition_2 = b < 0.5
    condition_3 = a+b>0.5
    
    return(condition_1&condition_2&condition_3*1)
```

```python
N_sim = 100000
list_of_results = []

for i in range(N_sim):
    list_of_results.append(simulation())
```

```python
sum(list_of_results)/float(N_sim)
```

# About 0.25 -> agrees with solution
